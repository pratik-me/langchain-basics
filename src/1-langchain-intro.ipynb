```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"]
```

```python
from langchain.agents import create_agent
from langchain_groq import ChatGroq    # Traditional way of initialization

llm = ChatGroq(model = "llama-3.3-70b-versatile", temperature = 0)

def get_weather(city: str) -> str:
    # Docstring will provide llm a brief about this tool
    '''Get the weather for a city.'''
    return f"The weather in {city} is sunny."

agent = create_agent(model = llm, tools = [get_weather], system_prompt = "You are a helpful assistant.")

agent
```

```python
### Run the agent
# If we only have to pass user message
# agent.invoke({"messages": "What is the weather of New York?"})

# Recommended method
message = agent.invoke({
    "messages": [ 
        {
            "role": "user",
            "content": "What is the weather of New York?"
        }
]})
```
