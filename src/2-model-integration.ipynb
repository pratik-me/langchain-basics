### Streaming and Batch

#### Streaming

Streaming is a built-in mechanism that allows developers to consume outputs from LLMs and complex workflows **incrementally**, rather than waiting for the entire execution to complete. By returning data token-by-token or step-by-step, streaming drastically reduces perceived latency and creates highly responsive, interactive interfaces—similar to the **live-typing effect** seen in ChatGPT or in other chatbots.

The stream method returns streamed outputs as iterators. We will use a loop to process each chunk in real-time.

```python
for chunk in model.stream("How does human civilization begins on earth? Briefly explain in 100 words"):
    print(chunk.text, end="\n", flush=True)
```

#### Batch

Batching a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing can be done in parallel

```python
responses = model.batch([
    "What's the difference between HTML and HTTP?",
    "What is the role of a UI/UX developer in software development?",
    "Name any 5 states of India and their capitals in json format"
])

for response in responses:
    print(response)
```

```python
responses = model.batch([
    "What's the difference between HTML and HTTP?",
    "What is the role of a UI/UX developer in software development?",
    "Name any 5 states of India and their capitals in json format"
], config={
    'max_concurrency': 5,      # Limit to 5 parallel calls
})

for response in responses:
    print(response)
```
