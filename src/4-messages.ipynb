### Messages

Messages are fundamental unit of context for models in Langchain. They represent the input and output of models, carrying both the content and metadata needed to represent the state of a conversation when interacting with an LLM. Messages are objects that contain:
+ Role: Identifies the message type (like system, user)
+ Content: Represents the actual content of the message (like text, image, audio, documents, etc)
+ Metadata: Optional fields such as response information, message IDs, and token usage

Langchain provides standard message type that works accross all the providers which ensure consistent behaviour regardless of which model is called.

```python
import os
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
model = init_chat_model("groq:llama-3.3-70b-versatile")
model
```

```python
model.invoke("What's the difference between Artificial Intelligence and Machine Learning?")   # This will return an AIMessage
```

#### Text Prompts

Text prompts are strings - ideal for straightforward generation tasks where you don't need to retain conversation history

Use text prompts when:
1. You have a single, standalone request
2. You don't need conversation history
3. You want minimal code complexity

```python
model.invoke("What's the difference between Artificial Intelligence and Machine Learning?")
```

#### Message Prompts

Message prompts can pass a list of messages to the model by providing a list of message objects.

Message types:
1. System message: Tells the model how to behave and provide context to interactions
2. Human message: Represents user input and interactions with the model
3. AI message: Responses generated by the model, including text content, tool calls, and metadata
4. Tool message: Represents the outputs of a tool call

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage, ToolMessage

messages = [
    SystemMessage("You are a helpful assistant. Your task is to give response under 100 words"),
    HumanMessage("What's the use of Artificial Intelligence in day to day life?")
]

response = model.invoke(messages)

response.content
```

```python
from langchain.messages import AIMessage, ToolMessage

ai_message = AIMessage(
    content = [],
    tool_calls = [{
        "name": "get_weather",
        "args": {"place": "New Delhi"},
        "id": "call_690",
    }]
)

tool_message = ToolMessage(
    content = "Sunny, 72 F",
    tool_call_id = "call_123"
)

messages = [
    HumanMessage("What's the weather in New Delhi?"),
    ai_message,
    tool_message,
]

response.invoke(messages)
```
